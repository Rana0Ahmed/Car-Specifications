## Import  needed modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('tableau-colorblind10')
```

```python
# read dataset using pandas
df = pd.read_csv(r'C:\Users\Rana Ahmed\Desktop\Cars Data1.csv')
df.head()
```

```python
# visualize the dataset
df.hist(bins=20, figsize=(10, 8))
plt.show()
```

## Data Cleaning

```python
df.shape
```

```python
# get information about dataset
print(df.info())
```

```python
# find all null values in the dataset
df.isnull().sum()
```

```python
# fill null values in 'Cylinders' column with the mean
df['Cylinders'] = df['Cylinders'].fillna(np.mean(df['Cylinders']))
```

```python
# drop all null values
df.dropna(axis=0, inplace=True)
```

```python
# check for null values
df.isnull().sum()
```

```python
# check values in 'Make' column
df['Make'].value_counts()
```

## Data Processing


```python
df.head()
```

```python
# group 'Invoice' based on 'Model'
df.groupby(['Model'])[['Invoice']].sum()
```

```python
# group 'Invoice' based on 'Horsepower'
df.groupby(['Horsepower'])[['Invoice']].sum()
```

```python
# group 'Horsepower' based on 'EngineSize'
df.groupby(['EngineSize'])[['Horsepower']].mean()
```

```python
df.groupby(['Model','Make'])[['Horsepower']].sum()
```

## Data Visualization

```python
# Invoice vs Horsepower
horsepower = df['Horsepower']
x_indices = np.arange(len(df['Invoice']))
plt.bar(x_indices , horsepower)

plt.title('Invoice vs Horsepower')
plt.xlabel('Invoice')
plt.ylabel('Horsepower')

plt.grid()
plt.show()
```
